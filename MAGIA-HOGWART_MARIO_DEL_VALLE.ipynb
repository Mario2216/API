```
import pandas as pd
import numpy as np
import random
import os

# 1. Definir los países y la estructura de los clústeres sintéticos
paises = [
    "Afghanistan", "Angola", "Benin", "Burundi", "Central African Republic", "Chad",
    "Congo, Dem. Rep.", "Eritrea", "Haiti", "Mozambique", "Niger", "Sierra Leone",
    "Somalia", "South Sudan", "Togo", "Yemen", "Uganda", "Sudan", "Mali", "Nigeria", # 20 Países Pobres
    "Brazil", "China", "Colombia", "Egypt", "India", "Indonesia", "Mexico", "Peru",
    "Russia", "South Africa", "Thailand", "Turkey", "Vietnam", "Chile", "Argentina",
    "Malaysia", "Philippines", "Ecuador", "Ukraine", "Poland", # 20 Países Medios
    "Australia", "Austria", "Canada", "Denmark", "France", "Germany", "Iceland",
    "Ireland", "Japan", "Kuwait", "Luxembourg", "Netherlands", "Norway", "Qatar",
    "Singapore", "South Korea", "Sweden", "Switzerland", "United Arab Emirates",
    "United States" # 20 Países Ricos
]
# Asegurar al menos 70 países
if len(paises) < 70:
    for i in range(len(paises), 70):
        paises.append(f"Pais_Sintetico_{i-len(paises)+1}")

data = {'country': paises[:70]}
df_sintetico = pd.DataFrame(data)

# 2. Simular las variables socioeconómicas por clúster (Poor, Middle, Rich)
n_poor = 25
n_middle = 25
n_rich = 20

# Crear una lista de rangos [min, max] para cada variable, segmentada por clúster
ranges = {
    # Variable: [Poor Range, Middle Range, Rich Range]
    'child_mort':   [[90, 150], [20, 80], [2, 10]],
    'exports':      [[5, 20], [20, 60], [40, 90]],
    'health':       [[0.5, 3], [3, 8], [8, 15]],
    'imports':      [[10, 30], [30, 70], [50, 100]],
    'income':       [[400, 1500], [1500, 15000], [15000, 60000]],
    'inflation':    [[8, 25], [3, 10], [0.5, 4]],
    'life_expec':   [[45, 60], [60, 75], [75, 85]],
    'total_fer':    [[4, 7], [2, 4], [1, 2.5]],
    'gdpp':         [[300, 800], [800, 15000], [15000, 100000]]
}

datos_generados = []

for i, country in enumerate(df_sintetico['country']):
    if i < n_poor:
        cluster_idx = 0  # Poor
    elif i < n_poor + n_middle:
        cluster_idx = 1  # Middle
    else:
        cluster_idx = 2  # Rich

    fila = {'country': country}
    for col, ranges_list in ranges.items():
        min_val, max_val = ranges_list[cluster_idx]
        if col in ['child_mort', 'total_fer', 'inflation']: # Variables con poca precisión
             fila[col] = np.random.uniform(min_val, max_val)
        else:
             # Generar enteros paragdpp y income si son grandes, o flotantes para otros
             if max_val > 1000:
                 fila[col] = random.randint(min_val, max_val)
             else:
                 fila[col] = np.random.uniform(min_val, max_val)

    datos_generados.append(fila)

df_final = pd.DataFrame(datos_generados)

# 3. Formatear y guardar como CSV
df_final = df_final[['country', 'child_mort', 'exports', 'health', 'imports', 'income', 'inflation', 'life_expec', 'total_fer', 'gdpp']]
df_final[['child_mort', 'health', 'imports', 'exports', 'inflation', 'total_fer']] = df_final[['child_mort', 'health', 'imports', 'exports', 'inflation', 'total_fer']].round(2)
df_final[['income', 'gdpp']] = df_final[['income', 'gdpp']].round(0).astype(int)

nombre_archivo = 'data.csv'
df_final.to_csv(nombre_archivo, index=False)

print(f"\n✅ Archivo '{nombre_archivo}' creado con éxito en tu sesión de Colab.")
print(f"Dataset generado con {len(df_final)} países y 10 características.")

# Muestra una parte de los datos generados para verificar
print("\nPrimeras 5 filas del dataset generado:")
print(df_final.head())
```

```
# ==============================================================================
# INFORME FINAL: AGRUPACIÓN DE PAÍSES PARA AYUDA INTERNACIONAL (ONG)
# ESTUDIANTE: Mario del Valle Álvarez
# CÓDIGO FINAL CORREGIDO
# ==============================================================================

# --- 0. LIBRERÍAS ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Librerías de Scikit-learn
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Librerías para el Mapa Mundi
import plotly.express as px
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

# ====================================================================
# !!! CARGA DE DATOS !!!
# ====================================================================
try:
    # Asume que el archivo 'data.csv' ha sido subido a Colab
    df = pd.read_csv('data.csv')

    # Establecer 'country' como índice
    if 'country' in df.columns:
        df.set_index('country', inplace=True)

    # Copia del DataFrame original para el análisis final
    df_original = df.copy()
    print("Carga de datos exitosa. Ejecutando análisis...")

except FileNotFoundError:
    print("ERROR FATAL: El archivo 'data.csv' NO fue encontrado en la sesión de Colab. Asegúrese de subirlo.")
    raise

# ====================================================================
## 1. Ejercicio 1: ANÁLISIS EXPLORATORIO DE DATOS (EDA)
# ====================================================================
print("\n--- 1.1 INSPECCIÓN ---")
print("Tamaño del dataset:", df.shape)
print("\nPrimeros registros:\n", df.head())
print("\nEstadísticos descriptivos:\n", df.describe().round(2))
print("\nTipos de características:\n")
df.info()

# Gráficos de Distribución
df_numerico = df.select_dtypes(include=['number'])
n_vars = len(df_numerico.columns)
fig, axes = plt.subplots(n_vars, 2, figsize=(15, 3.5 * n_vars))
plt.suptitle('1.2 Distribución de Variables (EDA)', y=1.02, fontsize=18)

for i, var in enumerate(df_numerico.columns):
    sns.histplot(df_numerico[var], kde=True, ax=axes[i, 0]).set_title(f'Densidad de {var}')
    sns.boxplot(y=df_numerico[var], ax=axes[i, 1]).set_title(f'Boxplot de {var}')

plt.tight_layout()
plt.show()


# ====================================================================
## 2. Ejercicio 2: EXTRACCIÓN DE CARACTERÍSTICAS
# ====================================================================

# Matriz de Correlación
plt.figure(figsize=(10, 8))
sns.heatmap(df_numerico.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('2.1 Matriz de Correlaciones', fontsize=16)
plt.show()

# Construcción de Indicadores Compuestos (Feature Engineering)
df_indicadores = pd.DataFrame(index=df.index)

# 1. INDICADOR SALUD (life_expec (+), health (+), -child_mort, -total_fer)
df_indicadores['Salud'] = (df['life_expec'] / df['life_expec'].mean()) + \
                         (df['health'] / df['health'].mean()) - \
                         (df['child_mort'] / df['child_mort'].mean()) - \
                         (df['total_fer'] / df['total_fer'].mean())

# 2. INDICADOR COMERCIO (exports, imports)
df_indicadores['Comercio'] = (df['exports'] / df['exports'].mean()) + \
                            (df['imports'] / df['imports'].mean())

# 3. INDICADOR FINANZAS (gdpp, income, -inflation)
df_indicadores['Finanzas'] = (df['gdpp'] / df['gdpp'].mean()) + \
                           (df['income'] / df['income'].mean()) - \
                           (df['inflation'] / df['inflation'].mean())


# Escalado de Indicadores (StandardScaler)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_indicadores)
df_scaled = pd.DataFrame(X_scaled, columns=df_indicadores.columns, index=df_indicadores.index)
print("\n--- 2.3 Datos Escalados (Muestra) ---")
print(df_scaled.head().round(2))


# ====================================================================
## 3. Ejercicio 3: ENTRENAMIENTO Y EVALUACIÓN DEL MODELO (BLOQUE CORREGIDO)
# ====================================================================

# 3.1 Ajuste de Hiperparámetros (k)
inertia = []
silhouette_scores = []
k_range_full = range(1, 11) # Rango completo: 1 a 10 (para Inercia)
k_range_silhouette = range(2, 11) # Rango para Silueta: 2 a 10

for k in k_range_full:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    labels = kmeans.fit_predict(X_scaled)
    inertia.append(kmeans.inertia_)
    if k >= 2: # El score de silueta requiere al menos dos clústeres
        silhouette_scores.append(silhouette_score(X_scaled, labels))

# Gráficos de ajuste
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
# Eje X: k_range_full (10 elementos) | Eje Y: inertia (10 elementos)
plt.plot(k_range_full, inertia, marker='o', color='blue')
plt.title('3.1 Método del Codo (Inercia)', fontsize=14)
plt.xlabel('Número de Clústeres (k)')
plt.ylabel('Inercia (SSE)')
plt.grid(True)

plt.subplot(1, 2, 2)
# Eje X: k_range_silhouette (9 elementos) | Eje Y: silhouette_scores (9 elementos)
plt.plot(k_range_silhouette, silhouette_scores, marker='o', color='red')
plt.title('3.1 Coeficiente de Silueta', fontsize=14)
plt.xlabel('Número de Clústeres (k)')
plt.ylabel('Puntuación de Silueta')
plt.grid(True)
plt.tight_layout()
plt.show()

k_optimo = 3 # Valor elegido tras análisis visual.

# 3.2 Entrenamiento Final y Visualización (PCA 2D)
kmeans_final = KMeans(n_clusters=k_optimo, random_state=42, n_init='auto')
df_original['Cluster'] = kmeans_final.fit_predict(X_scaled)

# Reducción de dimensionalidad con PCA
pca = PCA(n_components=2)
components = pca.fit_transform(X_scaled)
pca_df = pd.DataFrame(data = components, columns = ['PC1', 'PC2'], index=df_original.index)
pca_df['Cluster'] = df_original['Cluster']

plt.figure(figsize=(10, 7))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df,
                palette='viridis', style='Cluster', s=100)
plt.title(f'3.2 Clústeres de Países en 2D (PCA con {k_optimo} Grupos)', fontsize=16)
plt.xlabel(f'Componente Principal 1 ({round(pca.explained_variance_ratio_[0]*100, 2)}% Varianza)')
plt.ylabel(f'Componente Principal 2 ({round(pca.explained_variance_ratio_[1]*100, 2)}% Varianza)')
plt.grid(True)
plt.show()


# ====================================================================
## 4. Ejercicio 4: EVALUACIÓN DEL MODELO Y CONCLUSIONES DE NEGOCIO
# ====================================================================

# 4.1 Perfilado de Clústeres (Income y Child_mort)
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='Cluster', y='income', data=df_original)
plt.title('4.1 Ingreso Neto (Income) por Clúster', fontsize=16)
plt.grid(axis='y')

plt.subplot(1, 2, 2)
sns.boxplot(x='Cluster', y='child_mort', data=df_original)
plt.title('4.1 Mortalidad Infantil (child_mort) por Clúster', fontsize=16)
plt.grid(axis='y')

plt.tight_layout()
plt.show()

# 4.2 Asignación de Nivel de Ayuda (NOTA: La asignación depende de los boxplots generados)
# Se asume la siguiente estructura típica:
cluster_labels = {
    1: 'Mucha Ayuda (Prioridad Alta)',
    0: 'Ayuda Moderada',
    2: 'No Necesita Ayuda (Países Ricos)'
}
df_original['Nivel_Ayuda'] = df_original['Cluster'].map(cluster_labels)

print("\n--- 4.2 Perfil de los Clústeres por Nivel de Ayuda ---")
print(df_original.groupby('Nivel_Ayuda')[['income', 'child_mort', 'gdpp']].mean().sort_values(by='child_mort', ascending=False).round(2))


# 4.3 Mapa Mundi de Priorización
fig = px.choropleth(
    df_original.reset_index(),
    locations=df_original.index.name,
    locationmode='country names',
    color='Nivel_Ayuda',
    hover_name=df_original.index.name,
    color_discrete_map={
        'Mucha Ayuda (Prioridad Alta)': 'red',
        'Ayuda Moderada': 'orange',
        'No Necesita Ayuda (Países Ricos)': 'green'
    },
    title=f'4.3 Mapa de Priorización de Ayuda Internacional (k={k_optimo})'
)
fig.show()

# 4.4 Conclusiones y Mejoras
print("\n--- 4.4 Conclusiones Clave y Priorización ---")
cluster_prioritario = df_original[df_original['Nivel_Ayuda'] == 'Mucha Ayuda (Prioridad Alta)'].index.tolist()
print(f"La ONG 'Ayuda Internacional' debe priorizar a {len(cluster_prioritario)} países en la categoría 'Mucha Ayuda'.")
print("\nPaíses en la Máxima Prioridad (Receptores del 70-80% de los fondos):\n", cluster_prioritario)
```
