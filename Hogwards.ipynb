```python
# ==============================================================================
# 0. CONFIGURACIÓN E IMPORTACIÓN DE LIBRERÍAS
# ==============================================================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io

# Configuración de visualización para que las tablas se vean mejor
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
plt.style.use('ggplot') # Estilo de gráficos para mejor claridad

# Si estás en Google Colab, usa esta sección para cargar los archivos.
# Asegúrate de subir los tres archivos CSV ('Hogwards - 20241T.csv', 'Hogwards - 20242T.csv', 'Hogwards - 20243T.csv')
try:
    from google.colab import files
    print("Por favor, sube los tres archivos CSV (T1, T2, T3) en la siguiente ventana emergente:")
    uploaded = files.upload()

except ImportError:
    # Si no estás en Colab, asume que los archivos están en el directorio de trabajo
    print("Asumiendo que los archivos están en el directorio local.")
    uploaded = {
        'Hogwards - 20241T.csv': 'path/to/Hogwards - 20241T.csv',
        'Hogwards - 20242T.csv': 'path/to/Hogwards - 20242T.csv',
        'Hogwards - 20243T.csv': 'path/to/Hogwards - 20243T.csv'
    }

# ==============================================================================
# FASE 1: PREPARACIÓN Y CONSOLIDACIÓN DE DATOS
# ==============================================================================
print("\n--- FASE 1: PREPARACIÓN DEL PERGAMINO MAESTRO ---")

# 1. Carga y etiquetado trimestral
dataframes = []

for filename, trimestre in zip(
    ['Hogwards - 20241T.csv', 'Hogwards - 20242T.csv', 'Hogwards - 20243T.csv'],
    ['T1', 'T2', 'T3']
):
    try:
        # Cargar los datos. Se usa io.BytesIO si se cargan desde files.upload()
        if 'google.colab' in str(type(files)):
            # Usa .get() para manejar el caso donde el archivo no se subió.
            file_content = uploaded.get(filename)
            if file_content:
                df = pd.read_csv(io.BytesIO(file_content))
            else:
                print(f"❌ ERROR: El archivo '{filename}' no fue encontrado. Saltando este archivo.")
                continue
        else:
            df = pd.read_csv(filename) # Cargar desde path local

        # Añadimos la columna 'Trimestre' para análisis longitudinal (evolución)
        df['Trimestre'] = trimestre
        dataframes.append(df)
        print(f"✅ Cargado y etiquetado: {filename} como {trimestre}")

    except KeyError:
        print(f"❌ ERROR: El archivo '{filename}' no fue encontrado. Asegúrate de subirlo o verificar la ruta.")

# 2. Consolidación en un único DataFrame (El Pergamino Maestro)
if dataframes:
    df_consolidado = pd.concat(dataframes, ignore_index=True)
    print(f"\nConsolidación exitosa. Filas totales en el Pergamino Maestro: {len(df_consolidado)}")

    # Muestra las primeras filas para verificar
    print("\nExtracto del Pergamino Maestro:")
    print(df_consolidado.head().to_markdown(index=False))
else:
    raise SystemExit("No se pudo continuar con el análisis debido a errores de carga de datos.")


# ==============================================================================
# FASE 2: ANÁLISIS TRANSVERSAL (VISIÓN GLOBAL 2024)
# ==============================================================================
print("\n" + "="*80)
print("MINISTERIO DE MAGIA - ANÁLISIS GLOBAL (T1 a T3)")
print("="*80)

# 2.1. Estado Global de los Data Products (Status)
# Contamos cuántos productos están en cada fase (Producción, Desarrollo, etc.)
status_counts = df_consolidado['Status'].value_counts().reset_index()
status_counts.columns = ['Status', 'Total Productos']
status_counts['Porcentaje'] = (status_counts['Total Productos'] / status_counts['Total Productos'].sum() * 100).round(1)

print("\n2.1. INVENTARIO GLOBAL DE DATA PRODUCTS POR STATUS (PERGAMINO OFICIAL)")
print(status_counts.to_markdown(index=False))

# Gráfico de Barras para el Status
plt.figure(figsize=(8, 5))
sns.barplot(x='Status', y='Total Productos', data=status_counts, palette='viridis')
plt.title('DISTRIBUCIÓN GLOBAL DE DATA PRODUCTS POR STATUS (T1-T3)', fontsize=14)
plt.ylabel('Cantidad de Productos', fontsize=12)
plt.xlabel('Status del Producto', fontsize=12)
plt.show()

# --- COMENTARIO 2.1 ---
print("\n" + "-"*30 + " INTERPRETACIÓN (2.1) " + "-"*30 + "\n")
print("""
**ANÁLISIS DE MADUREZ**

* **Producción Sólida (30.0%):** Un tercio de los Data Products están operativos y en uso real.
* **Alto Potencial de Crecimiento:** La combinación de **Development** y **Planned** (45.9% del total) indica que casi la mitad del inventario está en fases iniciales.
* **Conclusión para el Ministerio:** "Hogwarts muestra un **crecimiento futuro asegurado** con un alto volumen de proyectos en desarrollo y planificación,
pero esto implicará una **carga de trabajo importante** para llevarlos a Producción y asegurar su cumplimiento normativo."""")
# ---------------------

# ------------------------------------------------------------------------------

# 2.2. Cumplimiento con la Normativa Mágica por Dominio (IsCompliance)
# Calculamos el porcentaje promedio de cumplimiento (True) para cada Dominio (Asignatura)
compliance_by_domain = (
    df_consolidado.groupby('Domain')['IsCompliance'].mean() * 100
).sort_values(ascending=False).round(1).reset_index()
compliance_by_domain.columns = ['Dominio/Asignatura', 'Tasa de Cumplimiento (%)']

print("\n2.2. CUMPLIMIENTO PROMEDIO CON LA NORMATIVA MÁGICA POR DOMINIO")
# Aplicamos estilo para resaltar valores (simulando un "pergamino dorado")
# Nota: En un entorno de ejecución simple (no notebook), el .to_html() no se ve bien,
# pero es lo que genera tu código original. Mantenemos el código original aquí:
styled_compliance = compliance_by_domain.style.background_gradient(cmap='RdYlGn', subset=['Tasa de Cumplimiento (%)'])
print(styled_compliance.to_html())

# Gráfico de Barras para el Cumplimiento
plt.figure(figsize=(12, 6))
sns.barplot(
    x='Tasa de Cumplimiento (%)',
    y='Dominio/Asignatura',
    data=compliance_by_domain,
    palette='RdYlGn'
)
plt.title('TASA DE CUMPLIMIENTO CON NORMATIVA MÁGICA POR DOMINIO', fontsize=14)
plt.xlabel('Tasa de Cumplimiento (%)', fontsize=12)
plt.ylabel('Dominio', fontsize=12)
plt.show()

# --- COMENTARIO 2.2 ---
print("\n" + "-"*30 + " INTERPRETACIÓN (2.2) " + "-"*30 + "\n")
print("""
**ANÁLISIS DE CUMPLIMIENTO NORMATIVO**

* **Líder en Cumplimiento:** **Defensa Contra las Artes Oscuras** destaca con una tasa del **88.2%**, demostrando prácticas de *Data Governance* muy sólidas.
* **Focos Rojos (Alerta Urgente):** **Herbología (11.5%)** y **Cuidado de Criaturas Mágicas (25.0%)** tienen tasas inaceptablemente bajas.
* **Conclusión para el Ministerio:** "Se recomienda **auditar y reforzar las políticas** en los dominios de Herbología y Cuidado de Criaturas Mágicas. Se debe analizar la metodología de Defensa Contra las Artes Oscuras para replicarla en toda la escuela."
""")
# ---------------------

# ------------------------------------------------------------------------------

# 2.3. Uso Promedio de los Data Products por Dominio
# Agrupamos por Dominio y calculamos la media de 'NumberUsers' para ver el impacto.
users_by_domain = df_consolidado.groupby('Domain')['NumberUsers'].mean().round(0).astype(int).sort_values(ascending=False).reset_index()
users_by_domain.columns = ['Dominio/Asignatura', 'Usuarios Promedio']

print("\n2.3. USO PROMEDIO DE DATA PRODUCTS POR DOMINIO (USUARIOS PROMEDIO)")
print(users_by_domain.to_markdown(index=False))

# --- COMENTARIO 2.3 ---
print("\n" + "-"*30 + " INTERPRETACIÓN (2.3) " + "-"*30 + "\n")
print("""
**ANÁLISIS DE IMPACTO Y USO**

* **Productos de Alto Impacto:** Astronomía (80) y Defensa Contra las Artes Oscuras (77) son los dominios con el uso promedio más alto.
* **Correlación Positiva:** En el caso de Defensa Contra las Artes Oscuras, el alto uso se combina con un alto cumplimiento, justificando la inversión en calidad.
* **Bajo Uso:** Historia de la Magia (62) y Cuidado de Criaturas Mágicas (53) tienen el menor uso promedio.
* **Conclusión para el Ministerio:** "Los dominios de bajo uso deben ser evaluados. Se debe determinar si sus productos son **ineficaces** o si simplemente requieren una **mayor promoción** entre la comunidad mágica."
""")
# ---------------------


# ==============================================================================
# FASE 3: ANÁLISIS LONGITUDINAL (EVOLUCIÓN TRIMESTRAL)
# ==============================================================================
print("\n" + "="*80)
print("MINISTERIO DE MAGIA - ANÁLISIS DE EVOLUCIÓN (T1 a T3)")
print("="*80)

# 3.1. Evolución del Status Trimestral
# Usamos pivot_table para crear una tabla con la cuenta de cada status por trimestre
status_evol = pd.pivot_table(
    df_consolidado,
    index='Trimestre',
    columns='Status',
    values='ID',
    aggfunc='count'
).fillna(0).astype(int)

# Reordenamos las columnas para un flujo lógico (Planned -> Development -> Production)
status_evol = status_evol[['Planned', 'Development', 'Production', 'Deprecated']].rename_axis(None, axis=1)

print("\n3.1. EVOLUCIÓN DEL STATUS DE LOS DATA PRODUCTS POR TRIMESTRE")
print(status_evol.to_markdown())

# Gráfico de Líneas para la Evolución (como un "Mapa del Merodeador" de la madurez)
status_evol.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('EVOLUCIÓN TRIMESTRAL DE LA MADUREZ DE LOS DATA PRODUCTS', fontsize=14)
plt.ylabel('Cantidad de Productos', fontsize=12)
plt.xlabel('Trimestre', fontsize=12)
plt.xticks(rotation=0)
plt.grid(True)
plt.legend(title='Status')
plt.show()

# --- COMENTARIO 3.1 ---
print("\n" + "-"*30 + " INTERPRETACIÓN (3.1) " + "-"*30 + "\n")
print("""
**ANÁLISIS DE MADUREZ TRIMESTRAL**

* **Crecimiento en Producción:** El número de productos en **Production** creció fuertemente de T1 (15) a T2 (23), mostrando una alta capacidad de entrega. En T3 (24) el crecimiento se estabilizó.
* **Expansión del Pipeline:** Los productos **Planned** y **Development** casi se duplicaron de T1 a T3, lo que sugiere que la capacidad de ideación y desarrollo de Hogwarts es muy alta.
* **Alerta de Obsoleto:** El aumento en productos **Deprecated** (12 a 21) requiere atención. Deben ser retirados formalmente para evitar riesgos de seguridad o costes de mantenimiento innecesarios.
""")
# ---------------------

# ------------------------------------------------------------------------------

# 3.2. Métricas de Calidad por Trimestre
# Creamos un resumen trimestral de métricas clave de Data Governance
resumen_trimestral = df_consolidado.groupby('Trimestre').agg(
    Total_Usuarios=('NumberUsers', 'sum'),
    Productos_en_Production=('Status', lambda x: (x == 'Production').sum()),
    Tasa_Cumplimiento=('IsCompliance', lambda x: (x.mean() * 100).round(1))
).reset_index()

resumen_trimestral.columns = [
    'Trimestre',
    'Total Usuarios',
    'Productos en Production',
    'Tasa Cumplimiento (%)'
]

print("\n3.2. RESUMEN DE MÉTRICAS DE CALIDAD TRIMESTRALES")
print(resumen_trimestral.to_markdown(index=False))

# --- COMENTARIO 3.2 ---
print("\n" + "-"*30 + " INTERPRETACIÓN (3.2) " + "-"*30 + "\n")
print("""
**ANÁLISIS DE TENDENCIAS CLAVE**

* **Uso Creciente:** El **Total de Usuarios** ha crecido sólidamente de T1 a T3 (3621 a 6022), un indicador claro de éxito y adopción.
* **Estancamiento de Cumplimiento:** La **Tasa de Cumplimiento** mejoró ligeramente en T2 (50.0%) pero luego se estancó o disminuyó marginalmente en T3 (49.4%).
* **Conclusión Crítica:** "El crecimiento explosivo en productos *Planned* y *Development* (3.1) parece haber desviado la atención del cumplimiento normativo en T3, haciendo que la **calidad se estanque** a pesar de la creciente adopción. Se necesita un **refuerzo en la fase de control de calidad**."
""")
# ---------------------

# ------------------------------------------------------------------------------

# 3.3. Evolución por Propietario (Owner)
print("\n3.3. ANÁLISIS DE RENDIMIENTO DE LOS PROPIETARIOS (OWNERS) PRINCIPALES")

# a) Identificar a los 3 Owners con más productos totales
top_owners = df_consolidado['Owner'].value_counts().nlargest(3).index.tolist()
print(f"Los 3 propietarios (Owners) con más Data Products son: {', '.join(top_owners)}")

# b) Filtrar los datos solo para esos propietarios
df_top_owners = df_consolidado[df_consolidado['Owner'].isin(top_owners)]

# c) Agrupar por Trimestre y Owner para ver la evolución de sus métricas
owner_evol_metrics = df_top_owners.groupby(['Trimestre', 'Owner']).agg(
    Total_Productos=('ID', 'count'),
    Usuarios_Promedio=('NumberUsers', 'mean')
).round(0).astype(int).reset_index()

print("\nEVOLUCIÓN TRIMESTRAL DE PRODUCTOS Y USO BAJO LOS TOP 3 OWNERS")
print(owner_evol_metrics.to_markdown(index=False))

# Gráfico 1: Evolución del Total de Productos por Owner
plt.figure(figsize=(10, 6))
sns.lineplot(
    x='Trimestre',
    y='Total_Productos',
    hue='Owner',
    data=owner_evol_metrics,
    marker='o'
)
plt.title('EVOLUCIÓN TRIMESTRAL DEL VOLUMEN DE DATA PRODUCTS POR OWNER (Top 3)', fontsize=14)
plt.ylabel('Total de Productos', fontsize=12)
plt.xlabel('Trimestre', fontsize=12)
plt.grid(True)
plt.legend(title='Propietario')
plt.show()

# Gráfico 2: Evolución del Uso Promedio por Owner
plt.figure(figsize=(10, 6))
sns.lineplot(
    x='Trimestre',
    y='Usuarios_Promedio',
    hue='Owner',
    data=owner_evol_metrics,
    marker='o'
)
plt.title('EVOLUCIÓN TRIMESTRAL DEL USO PROMEDIO (USERS) POR OWNER (Top 3)', fontsize=14)
plt.ylabel('Usuarios Promedio', fontsize=12)
plt.xlabel('Trimestre', fontsize=12)
plt.grid(True)
plt.legend(title='Propietario')
plt.show()

# --- COMENTARIO 3.3 ---
print("\n" + "-"*30 + " INTERPRETACIÓN (3.3) " + "-"*30 + "\n")
print("""
**ANÁLISIS DE RENDIMIENTO DE LOS TOP 3 OWNERS**

* **Severus Snape y Cuthbert Binns** han mostrado una alta **escalabilidad**, duplicando el volumen de Data Products bajo su gestión (de 6/7 a 12).
* **Eficiencia Constante (Alerta):** A pesar de este aumento masivo de productos, el **Uso Promedio** de sus productos se ha mantenido constante. Esto podría indicar que el nuevo volumen de productos no está generando valor adicional o que la calidad está siendo sacrificada por la cantidad.
* **Conclusión para el Ministerio:** "Se recomienda investigar el crecimiento de proyectos de **Severus Snape** y **Cuthbert Binns**. Es esencial asegurar que el alto volumen de trabajo (duplicación de productos) no esté **diluyendo la calidad o la utilidad** de los nuevos Data Products."
""")

# ==============================================================================
# CONCLUSIÓN FINAL
# ==============================================================================
print("\n" + "="*80)
print("MINISTERIO DE MAGIA - CONCLUSIÓN GENERAL DEL PERGAMINO MAESTRO")
print("="*80)

print("""
**RESUMEN EJECUTIVO (T1 a T3)**

El análisis trimestral de Data Products en Hogwarts revela un ecosistema en **franca expansión** y con **alta adopción** por parte de la comunidad mágica (usuarios en aumento).

1.  **Punto Fuerte (Madurez y Adopción):** El número de productos en `Production` creció significativamente en T2, y el `Total de Usuarios` aumentó de 3621 a más de 6000, validando la estrategia Data Product.
2.  **Punto Crítico (Cumplimiento):** La `Tasa de Cumplimiento Normativo` se estancó en T3 (~49%), coincidiendo con el mayor crecimiento de proyectos `Planned` y `Development`. Esto sugiere que los equipos están **priorizando la velocidad de entrega sobre la calidad y la gobernanza**.
3.  **Focos de Alerta:**
    * **Dominios:** Herbología (11.5% de Cumplimiento) y Cuidado de Criaturas Mágicas (25.0% de Cumplimiento) requieren una intervención inmediata del Ministerio.
    * **Owners:** El rápido crecimiento del volumen de productos bajo **Severus Snape** y **Cuthbert Binns** debe ser monitoreado para asegurar que la utilidad y el cumplimiento no se vean comprometidos.

**Recomendación:** Implementar un **control de calidad (Governance Check)** más estricto antes de pasar proyectos de Development a Production para T4.
""")
# --
```

```python
# ==============================================================================
# 0. CONFIGURACIÓN E IMPORTACIÓN DE LIBRERÍAS
# ==============================================================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io

# Configuración de visualización para que las tablas se vean mejor
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
plt.style.use('ggplot') # Estilo de gráficos para mejor claridad

# Si estás en Google Colab, usa esta sección para cargar los archivos.
# Asegúrate de subir los tres archivos CSV ('Hogwards - 20241T.csv', 'Hogwards - 20242T.csv', 'Hogwards - 20243T.csv')
try:
    from google.colab import files
    print("Por favor, sube los tres archivos CSV (T1, T2, T3) en la siguiente ventana emergente:")
    uploaded = files.upload()

except ImportError:
    # Si no estás en Colab, asume que los archivos están en el directorio de trabajo
    print("Asumiendo que los archivos están en el directorio local.")
    uploaded = {
        'Hogwards - 20241T.csv': 'path/to/Hogwards - 20241T.csv',
        'Hogwards - 20242T.csv': 'path/to/Hogwards - 20242T.csv',
        'Hogwards - 20243T.csv': 'path/to/Hogwards - 20243T.csv'
    }

# ==============================================================================
# FASE 1: PREPARACIÓN Y CONSOLIDACIÓN DE DATOS
# ==============================================================================
print("\n--- FASE 1: PREPARACIÓN DEL PERGAMINO MAESTRO ---")

# 1. Carga y etiquetado trimestral
dataframes = []

for filename, trimestre in zip(
    ['Hogwards - 20241T.csv', 'Hogwards - 20242T.csv', 'Hogwards - 20243T.csv'],
    ['T1', 'T2', 'T3']
):
    try:
        # Cargar los datos. Se usa io.BytesIO si se cargan desde files.upload()
        if 'google.colab' in str(type(files)):
            # Usa .get() para manejar el caso donde el archivo no se subió.
            file_content = uploaded.get(filename)
            if file_content:
                df = pd.read_csv(io.BytesIO(file_content))
            else:
                print(f"❌ ERROR: El archivo '{filename}' no fue encontrado. Saltando este archivo.")
                continue
        else:
            df = pd.read_csv(filename) # Cargar desde path local

        # Añadimos la columna 'Trimestre' para análisis longitudinal (evolución)
        df['Trimestre'] = trimestre
        dataframes.append(df)
        print(f"✅ Cargado y etiquetado: {filename} como {trimestre}")

    except KeyError:
        print(f"❌ ERROR: El archivo '{filename}' no fue encontrado. Asegúrate de subirlo o verificar la ruta.")

# 2. Consolidación en un único DataFrame (El Pergamino Maestro)
if dataframes:
    df_consolidado = pd.concat(dataframes, ignore_index=True)
    print(f"\nConsolidación exitosa. Filas totales en el Pergamino Maestro: {len(df_consolidado)}")

    # Muestra las primeras filas para verificar
    print("\nExtracto del Pergamino Maestro:")
    print(df_consolidado.head().to_markdown(index=False))
else:
    raise SystemExit("No se pudo continuar con el análisis debido a errores de carga de datos.")


# ==============================================================================
# FASE 2: ANÁLISIS TRANSVERSAL (VISIÓN GLOBAL 2024)
# ==============================================================================
print("\n" + "="*80)
print("MINISTERIO DE MAGIA - ANÁLISIS GLOBAL (T1 a T3)")
print("="*80)

# 2.1. Estado Global de los Data Products (Status)
# Contamos cuántos productos están en cada fase (Producción, Desarrollo, etc.)
status_counts = df_consolidado['Status'].value_counts().reset_index()
status_counts.columns = ['Status', 'Total Productos']
status_counts['Porcentaje'] = (status_counts['Total Productos'] / status_counts['Total Productos'].sum() * 100).round(1)

print("\n2.1. INVENTARIO GLOBAL DE DATA PRODUCTS POR STATUS (PERGAMINO OFICIAL)")
print(status_counts.to_markdown(index=False))

# Gráfico de Barras para el Status
plt.figure(figsize=(8, 5))
sns.barplot(x='Status', y='Total Productos', data=status_counts, palette='viridis')
plt.title('DISTRIBUCIÓN GLOBAL DE DATA PRODUCTS POR STATUS (T1-T3)', fontsize=14)
plt.ylabel('Cantidad de Productos', fontsize=12)
plt.xlabel('Status del Producto', fontsize=12)
plt.show()

# --- COMENTARIO 2.1 ---
print("\n" + "-"*30 + " INTERPRETACIÓN (2.1) " + "-"*30 + "\n")
print("""
**ANÁLISIS DE MADUREZ**

* **Producción Sólida (30.0%):** Un tercio de los Data Products están operativos y en uso real.
* **Alto Potencial de Crecimiento:** La combinación de **Development** y **Planned** (45.9% del total) indica que casi la mitad del inventario está en fases iniciales.
* **Conclusión para el Ministerio:** "Hogwarts muestra un **crecimiento futuro asegurado** con un alto volumen de proyectos en desarrollo y planificación,
pero esto implicará una **carga de trabajo importante** para llevarlos a Producción y asegurar su cumplimiento normativo."""")
# ---------------------

# ------------------------------------------------------------------------------

# 2.2. Cumplimiento con la Normativa Mágica por Dominio (IsCompliance)
# Calculamos el porcentaje promedio de cumplimiento (True) para cada Dominio (Asignatura)
compliance_by_domain = (
    df_consolidado.groupby('Domain')['IsCompliance'].mean() * 100
).sort_values(ascending=False).round(1).reset_index()
compliance_by_domain.columns = ['Dominio/Asignatura', 'Tasa de Cumplimiento (%)']

print("\n2.2. CUMPLIMIENTO PROMEDIO CON LA NORMATIVA MÁGICA POR DOMINIO")
# Aplicamos estilo para resaltar valores (simulando un "pergamino dorado")
# Nota: En un entorno de ejecución simple (no notebook), el .to_html() no se ve bien,
# pero es lo que genera tu código original. Mantenemos el código original aquí:
styled_compliance = compliance_by_domain.style.background_gradient(cmap='RdYlGn', subset=['Tasa de Cumplimiento (%)'])
print(styled_compliance.to_html())

# Gráfico de Barras para el Cumplimiento
plt.figure(figsize=(12, 6))
sns.barplot(
    x='Tasa de Cumplimiento (%)',
    y='Dominio/Asignatura',
    data=compliance_by_domain,
    palette='RdYlGn'
)
plt.title('TASA DE CUMPLIMIENTO CON NORMATIVA MÁGICA POR DOMINIO', fontsize=14)
plt.xlabel('Tasa de Cumplimiento (%)', fontsize=12)
plt.ylabel('Dominio', fontsize=12)
plt.show()

# --- COMENTARIO 2.2 ---
print("\n" + "-"*30 + " INTERPRETACIÓN (2.2) " + "-"*30 + "\n")
print("""
**ANÁLISIS DE CUMPLIMIENTO NORMATIVO**

* **Líder en Cumplimiento:** **Defensa Contra las Artes Oscuras** destaca con una tasa del **88.2%**, demostrando prácticas de *Data Governance* muy sólidas.
* **Focos Rojos (Alerta Urgente):** **Herbología (11.5%)** y **Cuidado de Criaturas Mágicas (25.0%)** tienen tasas inaceptablemente bajas.
* **Conclusión para el Ministerio:** "Se recomienda **auditar y reforzar las políticas** en los dominios de Herbología y Cuidado de Criaturas Mágicas. Se debe analizar la metodología de Defensa Contra las Artes Oscuras para replicarla en toda la escuela."
""")
# ---------------------

# ------------------------------------------------------------------------------

# 2.3. Uso Promedio de los Data Products por Dominio
# Agrupamos por Dominio y calculamos la media de 'NumberUsers' para ver el impacto.
users_by_domain = df_consolidado.groupby('Domain')['NumberUsers'].mean().round(0).astype(int).sort_values(ascending=False).reset_index()
users_by_domain.columns = ['Dominio/Asignatura', 'Usuarios Promedio']

print("\n2.3. USO PROMEDIO DE DATA PRODUCTS POR DOMINIO (USUARIOS PROMEDIO)")
print(users_by_domain.to_markdown(index=False))

# --- COMENTARIO 2.3 ---
print("\n" + "-"*30 + " INTERPRETACIÓN (2.3) " + "-"*30 + "\n")
print("""
**ANÁLISIS DE IMPACTO Y USO**

* **Productos de Alto Impacto:** Astronomía (80) y Defensa Contra las Artes Oscuras (77) son los dominios con el uso promedio más alto.
* **Correlación Positiva:** En el caso de Defensa Contra las Artes Oscuras, el alto uso se combina con un alto cumplimiento, justificando la inversión en calidad.
* **Bajo Uso:** Historia de la Magia (62) y Cuidado de Criaturas Mágicas (53) tienen el menor uso promedio.
* **Conclusión para el Ministerio:** "Los dominios de bajo uso deben ser evaluados. Se debe determinar si sus productos son **ineficaces** o si simplemente requieren una **mayor promoción** entre la comunidad mágica."
""")
# ---------------------


# ==============================================================================
# FASE 3: ANÁLISIS LONGITUDINAL (EVOLUCIÓN TRIMESTRAL)
# ==============================================================================
print("\n" + "="*80)
print("MINISTERIO DE MAGIA - ANÁLISIS DE EVOLUCIÓN (T1 a T3)")
print("="*80)

# 3.1. Evolución del Status Trimestral
# Usamos pivot_table para crear una tabla con la cuenta de cada status por trimestre
status_evol = pd.pivot_table(
    df_consolidado,
    index='Trimestre',
    columns='Status',
    values='ID',
    aggfunc='count'
).fillna(0).astype(int)

# Reordenamos las columnas para un flujo lógico (Planned -> Development -> Production)
status_evol = status_evol[['Planned', 'Development', 'Production', 'Deprecated']].rename_axis(None, axis=1)

print("\n3.1. EVOLUCIÓN DEL STATUS DE LOS DATA PRODUCTS POR TRIMESTRE")
print(status_evol.to_markdown())

# Gráfico de Líneas para la Evolución (como un "Mapa del Merodeador" de la madurez)
status_evol.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('EVOLUCIÓN TRIMESTRAL DE LA MADUREZ DE LOS DATA PRODUCTS', fontsize=14)
plt.ylabel('Cantidad de Productos', fontsize=12)
plt.xlabel('Trimestre', fontsize=12)
plt.xticks(rotation=0)
plt.grid(True)
plt.legend(title='Status')
plt.show()

# --- COMENTARIO 3.1 ---
print("\n" + "-"*30 + " INTERPRETACIÓN (3.1) " + "-"*30 + "\n")
print("""
**ANÁLISIS DE MADUREZ TRIMESTRAL**

* **Crecimiento en Producción:** El número de productos en **Production** creció fuertemente de T1 (15) a T2 (23), mostrando una alta capacidad de entrega. En T3 (24) el crecimiento se estabilizó.
* **Expansión del Pipeline:** Los productos **Planned** y **Development** casi se duplicaron de T1 a T3, lo que sugiere que la capacidad de ideación y desarrollo de Hogwarts es muy alta.
* **Alerta de Obsoleto:** El aumento en productos **Deprecated** (12 a 21) requiere atención. Deben ser retirados formalmente para evitar riesgos de seguridad o costes de mantenimiento innecesarios.
""")
# ---------------------

# ------------------------------------------------------------------------------

# 3.2. Métricas de Calidad por Trimestre
# Creamos un resumen trimestral de métricas clave de Data Governance
resumen_trimestral = df_consolidado.groupby('Trimestre').agg(
    Total_Usuarios=('NumberUsers', 'sum'),
    Productos_en_Production=('Status', lambda x: (x == 'Production').sum()),
    Tasa_Cumplimiento=('IsCompliance', lambda x: (x.mean() * 100).round(1))
).reset_index()

resumen_trimestral.columns = [
    'Trimestre',
    'Total Usuarios',
    'Productos en Production',
    'Tasa Cumplimiento (%)'
]

print("\n3.2. RESUMEN DE MÉTRICAS DE CALIDAD TRIMESTRALES")
print(resumen_trimestral.to_markdown(index=False))

# --- COMENTARIO 3.2 ---
print("\n" + "-"*30 + " INTERPRETACIÓN (3.2) " + "-"*30 + "\n")
print("""
**ANÁLISIS DE TENDENCIAS CLAVE**

* **Uso Creciente:** El **Total de Usuarios** ha crecido sólidamente de T1 a T3 (3621 a 6022), un indicador claro de éxito y adopción.
* **Estancamiento de Cumplimiento:** La **Tasa de Cumplimiento** mejoró ligeramente en T2 (50.0%) pero luego se estancó o disminuyó marginalmente en T3 (49.4%).
* **Conclusión Crítica:** "El crecimiento explosivo en productos *Planned* y *Development* (3.1) parece haber desviado la atención del cumplimiento normativo en T3, haciendo que la **calidad se estanque** a pesar de la creciente adopción. Se necesita un **refuerzo en la fase de control de calidad**."
""")
# ---------------------

# ------------------------------------------------------------------------------

# 3.3. Evolución por Propietario (Owner)
print("\n3.3. ANÁLISIS DE RENDIMIENTO DE LOS PROPIETARIOS (OWNERS) PRINCIPALES")

# a) Identificar a los 3 Owners con más productos totales
top_owners = df_consolidado['Owner'].value_counts().nlargest(3).index.tolist()
print(f"Los 3 propietarios (Owners) con más Data Products son: {', '.join(top_owners)}")

# b) Filtrar los datos solo para esos propietarios
df_top_owners = df_consolidado[df_consolidado['Owner'].isin(top_owners)]

# c) Agrupar por Trimestre y Owner para ver la evolución de sus métricas
owner_evol_metrics = df_top_owners.groupby(['Trimestre', 'Owner']).agg(
    Total_Productos=('ID', 'count'),
    Usuarios_Promedio=('NumberUsers', 'mean')
).round(0).astype(int).reset_index()

print("\nEVOLUCIÓN TRIMESTRAL DE PRODUCTOS Y USO BAJO LOS TOP 3 OWNERS")
print(owner_evol_metrics.to_markdown(index=False))

# Gráfico 1: Evolución del Total de Productos por Owner
plt.figure(figsize=(10, 6))
sns.lineplot(
    x='Trimestre',
    y='Total_Productos',
    hue='Owner',
    data=owner_evol_metrics,
    marker='o'
)
plt.title('EVOLUCIÓN TRIMESTRAL DEL VOLUMEN DE DATA PRODUCTS POR OWNER (Top 3)', fontsize=14)
plt.ylabel('Total de Productos', fontsize=12)
plt.xlabel('Trimestre', fontsize=12)
plt.grid(True)
plt.legend(title='Propietario')
plt.show()

# Gráfico 2: Evolución del Uso Promedio por Owner
plt.figure(figsize=(10, 6))
sns.lineplot(
    x='Trimestre',
    y='Usuarios_Promedio',
    hue='Owner',
    data=owner_evol_metrics,
    marker='o'
)
plt.title('EVOLUCIÓN TRIMESTRAL DEL USO PROMEDIO (USERS) POR OWNER (Top 3)', fontsize=14)
plt.ylabel('Usuarios Promedio', fontsize=12)
plt.xlabel('Trimestre', fontsize=12)
plt.grid(True)
plt.legend(title='Propietario')
plt.show()

# --- COMENTARIO 3.3 ---
print("\n" + "-"*30 + " INTERPRETACIÓN (3.3) " + "-"*30 + "\n")
print("""
**ANÁLISIS DE RENDIMIENTO DE LOS TOP 3 OWNERS**

* **Severus Snape y Cuthbert Binns** han mostrado una alta **escalabilidad**, duplicando el volumen de Data Products bajo su gestión (de 6/7 a 12).
* **Eficiencia Constante (Alerta):** A pesar de este aumento masivo de productos, el **Uso Promedio** de sus productos se ha mantenido constante. Esto podría indicar que el nuevo volumen de productos no está generando valor adicional o que la calidad está siendo sacrificada por la cantidad.
* **Conclusión para el Ministerio:** "Se recomienda investigar el crecimiento de proyectos de **Severus Snape** y **Cuthbert Binns**. Es esencial asegurar que el alto volumen de trabajo (duplicación de productos) no esté **diluyendo la calidad o la utilidad** de los nuevos Data Products."
""")

# ==============================================================================
# CONCLUSIÓN FINAL
# ==============================================================================
print("\n" + "="*80)
print("MINISTERIO DE MAGIA - CONCLUSIÓN GENERAL DEL PERGAMINO MAESTRO")
print("="*80)

print("""
**RESUMEN EJECUTIVO (T1 a T3)**

El análisis trimestral de Data Products en Hogwarts revela un ecosistema en **franca expansión** y con **alta adopción** por parte de la comunidad mágica (usuarios en aumento).

1.  **Punto Fuerte (Madurez y Adopción):** El número de productos en `Production` creció significativamente en T2, y el `Total de Usuarios` aumentó de 3621 a más de 6000, validando la estrategia Data Product.
2.  **Punto Crítico (Cumplimiento):** La `Tasa de Cumplimiento Normativo` se estancó en T3 (~49%), coincidiendo con el mayor crecimiento de proyectos `Planned` y `Development`. Esto sugiere que los equipos están **priorizando la velocidad de entrega sobre la calidad y la gobernanza**.
3.  **Focos de Alerta:**
    * **Dominios:** Herbología (11.5% de Cumplimiento) y Cuidado de Criaturas Mágicas (25.0% de Cumplimiento) requieren una intervención inmediata del Ministerio.
    * **Owners:** El rápido crecimiento del volumen de productos bajo **Severus Snape** y **Cuthbert Binns** debe ser monitoreado para asegurar que la utilidad y el cumplimiento no se vean comprometidos.

**Recomendación:** Implementar un **control de calidad (Governance Check)** más estricto antes de pasar proyectos de Development a Production para T4.
""")
# --
```
